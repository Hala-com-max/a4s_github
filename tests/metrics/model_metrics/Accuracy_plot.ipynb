```python
### Cell to Inspect CSV Contents ###

import pandas as pd
import os
import matplotlib.pyplot as plt

MEASURES_PATH = "./tests/data/measures/accuracy.csv"

if os.path.exists(MEASURES_PATH):
    print(f"Contents of {MEASURES_PATH} (first 10 rows):")
    try:
        # Read the CSV file
        df_check = pd.read_csv(MEASURES_PATH)
        
        # Filter for accuracy rows
        df_accuracy = df_check[df_check['name'] == 'accuracy']
        
        # Calculate and print the mean to confirm the data type
        mean_score = df_accuracy['score'].mean()
        
        print(df_accuracy.head().to_markdown(index=False))
        print(f"\nCalculated Mean Score in this file: {mean_score:.4f}")
        
    except Exception as e:
        print(f"Error reading or processing CSV: {e}")
else:
    print(f"Error: File not found at {MEASURES_PATH}. Run 'uv run pytest...' first.")
```
